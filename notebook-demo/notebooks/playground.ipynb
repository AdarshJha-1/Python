```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("../images/cat.jpeg")
img

```

```python

plt.title("Image")
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
arr = np.array(img)
arr
```

```python
arr[0, 0]
```

```python
# inverting the color
invert = 255 - arr
plt.title("Inverted Image")
plt.imshow(invert)
plt.show()
```

```python
# Brightness
bright = np.clip(arr+10, 0, 255)
plt.imshow(bright)
```

```python
# Darken
dark = np.clip(arr-20, 0, 255)
plt.imshow(dark)
```

```python
# Grayscale
gray = arr.mean(axis=2)
plt.imshow(gray,cmap="gray")
```

```python
# Crop
crop = arr[100:500,100:600]
plt.imshow(crop)
```

```python
# Flip Horizontal
horizontal = arr[:,::-1]
plt.imshow(horizontal)
```

```python
# Flip Vertical
vertical = arr[::-1,:]
plt.imshow(vertical)
plt.axis("off")
```

```python
# Rotate 90-Degree
plt.imshow(np.rot90(arr))
plt.axis("off")
plt.show()
```

```python
# Blur
output = np.zeros_like(arr)

for i in range(1, arr.shape[0]-1):
    for j in range(1, arr.shape[1]-1):
        region = arr[i-1:i+2, j-1:j+2]
        output[i, j] = region.mean(axis=(0, 1))

plt.title("Blurred Image")
plt.imshow(output)
plt.axis("off")
plt.show()
```
